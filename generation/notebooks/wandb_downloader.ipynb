```python
# walk the subfolders under drive_folder and replace the prefix "X_" with "step" for image files
# update path and uncomment
drive_folder = "/content/drive/MyDrive/wandb_images/pmnist_v7_A100_2ksteps"

import os
import re

def replace_prefix_in_subfolders(drive_folder):
  for root, dirs, files in os.walk(drive_folder):
    for file in files:
      if file.startswith(('0_', '1_', '2_', '3_', '4_', '5_', '6_', '7_', '8_', '9_')):
        old_path = os.path.join(root, file)
        new_file = re.sub(r"^[0-9]_", "step", file)  # Replace prefix with "step"
        new_path = os.path.join(root, new_file)
        os.rename(old_path, new_path)
        print(f"Renamed: {old_path} to {new_path}")

replace_prefix_in_subfolders(drive_folder)
```

```python
import os
import shutil
import re

# walk the subfolders of drive_folder and organize the images by prefix stepX into a subfolder named StepX, where X is between 0 to 9999
# update path and uncomment (to avoid accidental multiple updates to same folder)
drive_folder = "/content/drive/MyDrive/wandb_images/pmnist_v7_A100_2ksteps"

def organize_images(drive_folder):
  for root, dirs, files in os.walk(drive_folder):
    for file in files:
      # Check if the file starts with "step" followed by numbers
      match = re.match(r"step(\d+)", file)
      if match:
        prefix = match.group(0)  # Extract the prefix (e.g., "step123")

        # Create a subfolder with the prefix name if it doesn't exist
        subfolder_path = os.path.join(root, prefix)
        os.makedirs(subfolder_path, exist_ok=True)

        # Move the file to the subfolder
        old_path = os.path.join(root, file)
        new_path = os.path.join(subfolder_path, file)
        shutil.move(old_path, new_path)
        print(f"Moved: {old_path} to {new_path}")

# Call the function to organize images in your drive_folder
organize_images(drive_folder)
```
